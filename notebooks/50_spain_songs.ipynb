```python
data = pd.read_csv("/Users/PUSHPAK/OneDrive/Desktop/Spain_Project/data/normalized_songs.csv")
data.head()
```

```python
data.groupby('date').size()
```

```python
#Check Playlist Size per Day, Top-50 playlist must contain exactly 50 songs per day.
data.groupby('date').size().unique()
```

```python
data = data.groupby('date').filter(lambda x: len(x) == 50)
```

```python
data.groupby('date').size().unique()
```

```python
#Validate Position Column, Position should only be 1–50.
data['position'].describe()
```

```python
data = data[(data['position'] >= 1) & (data['position'] <= 50)]
```

```python
data.head()
```

```python
#Clean Album Type Column
data['album_type'].unique()
```

```python
data['album_type'] = data['album_type'].str.lower().str.strip()
```

```python
data['is_explicit'] = data['is_explicit'].astype(bool)
```

```python
data
```

```python
data.info()
```

```python
data = data.reset_index(drop=True)
```

```python
#Create Unique Song ID
data['song_id'] = data['song'] + "_" + data['artist']
```

```python
data.head()
```

```python
#Sort Dataset
data = data.sort_values(by=['song_id','date'])
```

```python
data = data.drop(columns=['album_cover_url'])
```

```python
data.head()
```

```python
data.to_csv("/Users/PUSHPAK/OneDrive/Desktop/Spain_Project/data/normalized_2_songs.csv", index=False)
```
